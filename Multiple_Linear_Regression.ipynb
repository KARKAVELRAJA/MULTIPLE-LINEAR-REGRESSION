## ***IMPORTING THE LIBRARIES***

```python
import numpy as np
import pandas as pd
```

## ***LOADING THE DATASET***

```python
data = pd.read_csv("Marketing_Data.csv")
```

## ***SUMMARIZING THE DATASET***

```python
print(data.shape)
print(data.head())
```

## ***SEGREGATING THE DATASET INTO INPUT AND OUTPUT***

```python
input  = data.iloc[:,:-1].values
print(input)
output = data.iloc[:,-1].values
print(output)
```

## ***VISUALIZING THE DATASET***

```python
import matplotlib.pyplot as plt

plt.scatter(input[:,:1],output,color="Red")
plt.scatter(input[:,1:2],output,color="Green",marker="*")
plt.scatter(input[:,2:],output,marker="^")
plt.xlabel("BUDGET FOR SOCIAL MEDIA ADVERTISING")
plt.ylabel("SALES")
plt.legend(["YOUTUBE","FACEBOOK","NEWSPAPER"])
```

## ***TRAINING THE DATASET***

```python
from sklearn.linear_model import LinearRegression

Multiple_Linear_Regression_Model = LinearRegression()

Multiple_Linear_Regression_Model.fit(input,output)
```

## ***PREDICTING THE OUTPUT USING THE TRAINED MODEL***

```python
X1 = 246.00
X2 = 54.12
X3 = 23.52

print("BASED ON THE GIVEN BUDGET FOR SOCIAL MEDIA ADVERTISING, THE SALES IS ",Multiple_Linear_Regression_Model.predict([[X1,X2,X3]]))
```

## ***CHECKING THE PREDICTED OUTPUT USING  Y=(M1X1)+(M2X2)+(M3X3)+C***

```python
M1,M2,M3 = Multiple_Linear_Regression_Model.coef_
C        = Multiple_Linear_Regression_Model.intercept_

Y        = (M1*X1)+(M2*X2)+(M3*X3)+C


print("BASED ON THE GIVEN BUDGET FOR SOCIAL MEDIA ADVERTISING, THE SALES IS ",Y)
```
